# 二円と接線

円$C_1 : x^2+y^2=1$と円$C_2 : (x-2)^2+(y-4)^2=5$に点Pから接線を引く。Pから$C_1$までの距離と$C_2$までの距離の比が$1:2$になるとする。このとき，Pの軌跡を求めよ。


```julia
using Plots
```

```julia
gr()
```

```julia
θ=range(0,2π,length=1000)
plot(cos.(θ),sin.(θ),aspect_ratio=1)
plot!(sqrt.(5)*cos.(θ).+2,sqrt.(5)*sin.(θ).+4,aspect_ratio=1)
plot!(sqrt.(77)/3*cos.(θ).-2/3,sqrt.(77)/3*sin.(θ).-4/3,aspect_ratio=1)
```
