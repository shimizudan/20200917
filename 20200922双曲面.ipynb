## 双曲面

2点P$(0,1,1)$，Q$(1,0,-1)$を通る直線を$z$軸の周りに回転んしてできる曲面は？

直線PQのパラメータ表示は

$$\begin{pmatrix}
x\\
y\\
z\\
\end{pmatrix}=
\begin{pmatrix}
t\\
t-1\\
-2t+1\\
\end{pmatrix}$$

$z=-2t+1=k$と置き換えると，


$$\begin{pmatrix}
x\\
y\\
z\\
\end{pmatrix}=
\begin{pmatrix}
\frac{1-k}2\\
-\frac{k+1}2\\
k\\
\end{pmatrix}$$

この点を平面$z=0$と平行に回転させる。$\sqrt{\left(\frac{1-k}2\right)^2+\left(-\frac{k+1}2\right)^2}=\sqrt{\frac{k^2+1}2}$より，


$$\begin{pmatrix}
x\\
y\\
z\\
\end{pmatrix}=
\begin{pmatrix}
\sqrt{\frac{k^2+1}2}\cos\theta\\\
\sqrt{\frac{k^2+1}2}\sin\theta\\
k\\
\end{pmatrix}$$

２式より，$k$，$\theta$を消去する。
$$x^2+y^2=\frac{k^2+1}2=\frac{z^2+1}2$$

$$\therefore 2x^2+2y^2-z^2=1$$


$$\therefore z=\pm\sqrt{2x^2+2y^2-1}$$

```julia
x=-3.0:0.2:3.0
y=-3.0:0.2:3.0

function f(X,Y)
   
if 2*X^2+2*Y^2-1 < 0
        0.0
        else sqrt(2*X^2+2*Y^2-1)
    end
end

z=[f(X,Y) for X in x, Y in y]'
wireframe(x,y,f) 

```

```julia
x=-3.0:0.2:3.0
y=-3.0:0.2:3.0

function f(X,Y)
   
if 2*X^2+2*Y^2-1 < 0
        0.0
        else -1*sqrt(2*X^2+2*Y^2-1)
    end
end

z=[f(X,Y) for X in x, Y in y]'
wireframe(x,y,f) 
```
